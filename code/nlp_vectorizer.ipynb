```python
# Imports 
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
import pandas as pd

# Sample text data (document)

document = [
    "The sun set behind the mountains, casting a golden glow over the valley.",
    "She picked up the book and began to read, losing herself in the story.",
    "The cat stretched lazily on the windowsill, enjoying the warmth of the sun.",
    "He ran as fast as he could, determined to reach the finish line first.",
    "The rain pattered softly against the roof, creating a soothing melody.",
    "They gathered around the campfire, sharing stories and roasting marshmallows.",
    "The city skyline glittered with lights, a testament to human ingenuity.",
    "She laughed heartily, her joy infectious to everyone around her.",
    "The old oak tree stood tall and proud, its branches spreading wide.",
    "He meticulously painted the canvas, each stroke a reflection of his emotions."
]


#Creating a CountVectorizer object to count word frequencies
count_vectorizer = CountVectorizer()

#generating the Term-Document Matrix (TDM) / Document-Term Matrix (DTM)
# the matrix contains frequencies of each word in the document
# Here \, the TDM and DTM are the same; the naming deoends on the context
term_document_matrix = count_vectorizer.fit_transform(document)

#Covert the matrix to a DataFrame for better visualization
tdm_df = pd.DataFrame(term_document_matrix.toarray(), columns = count_vectorizer.get_feature_names_out())

# Print the Term-Document Matrix
print("Term-Document Matrix (TDM) / Document-Term Matrix (DTM")
print(tdm_df)
print("/n")

# Explantion:
#Rows represent documents, and columns represents term (words).
# Each cell (i , j) contains the frequency of the word j in document i

# Applying TF-IDF transformation
tfidf_transformer = TfidfTransformer()
tfidf_matrix = tfidf_transformer.fit_transform(term_document_matrix)

# Convert the TF-IDF matrix to a DataFrame for better visualization
tfidf =  pd.DataFrame(tfidf_matrix.toarray(), columns = count_vectorizer.get_feature_names_out())

# Print the TF-IDF Matrix
print("TF-IDF Matrix")
print(tfidf)
```

```python
# Explanation:
#TF-IDF (Term Frequency - Inverse Document Frequency) normalizes the word frequency
# by accounting for the frequency of the word word across all documents.
# This helps in reducing the impact of common words and highlighting important words
```
