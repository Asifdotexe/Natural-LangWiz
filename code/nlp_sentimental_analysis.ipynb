Using VADER

```python
import re
import nltk 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
```

```python
nltk.download('vader_lexicon')
```

```python
df = pd.read_csv(r'..\data\preprocessed_amazon_review.csv')
df.info()
```

```python
df.head()
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
def clean_text(text):
    if text is None:
        text = ''
    # Remove non-textual characters
    pattern = re.compile(r'[^\w\s]')
    cleaned_text = pattern.sub('', str(text))
    return cleaned_text

def get_sentiment_scores(text):
    cleaned_text = clean_text(text)
    return sia.polarity_scores(cleaned_text)
```

```python
df['sentiment_scores'] = df['reviewText'].apply(get_sentiment_scores)
```

```python
df.head()
```

```python
sentiment_df = pd.json_normalize(df['sentiment_scores'])
sentiment_df
```

```python
df = pd.concat([df,sentiment_df])
df.drop(columns=['sentiment_scores'], inplace=True)
```
