```python
import requests
import pandas as pd
```

# GET Method

```python
url = "https://jsonplaceholder.typicode.com/posts"
```

```python
response = requests.get(url)
print(response)
```

```python
if response.status_code == 200:
    data = response.json()
    df = pd.json_normalize(data)
```
